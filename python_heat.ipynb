```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# 给定点集
points = [(0,0.5),(0.9,1),(1.5,2.9),(2.9,2),(3.5,3.5)]

# 提取 x 和 y 坐标
x = [p[0] for p in points]
y = [p[1] for p in points]

# 创建插值函数
interp_func = interp1d(x, y, kind='cubic')

# 生成插值点
x_interp = np.linspace(min(x), max(x), num=100)
y_interp = interp_func(x_interp)

# 绘制插值曲线图
plt.plot(x_interp, y_interp, label='Interpolation')
plt.scatter(x, y, color='red', label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolation Curve')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# 给定的点集
points = [(0.9, 1), (1.5, 2.9), (2.9, 2), (3.5, 3.5)]
x = [point[0] for point in points]
y = [point[1] for point in points]

# 使用拉格朗日插值方法
poly = lagrange(x, y)

# 生成插值曲线上的更多点
x_interp = np.linspace(min(x), max(x), 100)
y_interp = poly(x_interp)

# 绘制原始点集和插值曲线
plt.plot(x, y, 'ro', label='Data Points')
plt.plot(x_interp, y_interp, 'b-', label='Interpolated Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interpolated Curve using Lagrange Interpolation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# 原始数据点
data_points = [(0.9, 1), (1.5, 2.9), (2.9, 2), (3.5, 3.5)]
x, y = zip(*data_points)

# 创建插值函数
interp_func = interp1d(x, y, kind='cubic')

# 生成更密集的数据点
x_interp = np.linspace(min(x), max(x), num=100)
y_interp = interp_func(x_interp)

# 绘制原始数据点和插值曲线
plt.plot(x, y, 'o', label='原始数据点')
plt.plot(x_interp, y_interp, label='插值曲线')

# 添加第二组数据点
data_points2 = [(0, 0.5), (0.9, 1)]
x2, y2 = zip(*data_points2)

# 创建第二组数据点的插值函数
interp_func2 = interp1d(x2, y2, kind='cubic')

# 生成第二组数据点的更密集数据
x_interp2 = np.linspace(min(x2), max(x2), num=100)
y_interp2 = interp_func2(x_interp2)

# 绘制第二组数据点和插值曲线
plt.plot(x2, y2, 'o', label='第二组数据点')
plt.plot(x_interp2, y_interp2, label='第二组插值曲线')

# 设置图例和标题
plt.legend()
plt.title('插值曲线图')

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# 数据点
data_points = [(0, 0.5), (0.9, 1)]
x, y = zip(*data_points)

# 计算边界条件
slope = 0  # 在(0, 0.5)点处的导数为零
end_slope = [slope, slope]

# 创建插值函数
interp_func = CubicSpline(x, y, bc_type=((1, end_slope), (1, end_slope)))

# 生成更密集的数据点
x_interp = np.linspace(min(x), max(x), num=100)
y_interp = interp_func(x_interp)

# 绘制数据点和插值曲线
plt.plot(x, y, 'o', label='数据点')
plt.plot(x_interp, y_interp, label='插值曲线')

# 设置图例和标题
plt.legend()
plt.title('插值曲线图')

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BPoly

# 数据点
data_points = [(0, 0.5), (0.9, 1)]
x, y = zip(*data_points)

# 在(0, 0.5)点处的导数为零
derivatives = np.zeros_like(x)

# 创建插值函数
interp_func = BPoly.from_derivatives(x, derivatives)

# 生成更密集的数据点
x_interp = np.linspace(min(x), max(x), num=100)
y_interp = interp_func(x_interp)

# 绘制数据点和插值曲线
plt.plot(x, y, 'o', label='数据点')
plt.plot(x_interp, y_interp, label='插值曲线')

# 设置图例和标题
plt.legend()
plt.title('插值曲线图')

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt

# 温度数据
temperature = [20, 30, 40, 50, 60, 70, 80, 90, 100]  # 单位：摄氏度

# 沸腾状态
boiling_state = ['起始阶段', '核态沸腾', '次临界沸腾', '临界沸腾', '过热沸腾']

# 沸腾曲线数据
boiling_curve = [0, 1, 2, 3, 4]

# 绘制沸腾曲线
plt.plot(temperature, boiling_curve, marker='o')

# 添加标签
for i in range(len(temperature)):
    plt.text(temperature[i], boiling_curve[i], boiling_state[i], ha='center', va='bottom')

# 设置图形标题和轴标签
plt.title('Water Boiling Curve')
plt.xlabel('Temperature (°C)')
plt.ylabel('Boiling State')

# 显示图形
plt.show()
```
